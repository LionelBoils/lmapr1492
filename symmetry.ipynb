```python
import matplotlib.pyplot as plt
import matplotlib.image as img
from pymatgen.ext.matproj import MPRester
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
with MPRester ("Js0y7wlKAQ1kbIXr") as m:
    #telechargement
    structure = m.get_structure_by_material_id("mp-11178")
    #structure
s = SpacegroupAnalyzer(structure)
#structure conventionelle
simple_s = s.get_conventional_standard_structure()
#spacegroup
spacegroup= SpacegroupAnalyzer(simple_s)
#coordonnees sites
site = simple_s.sites
#renvoie la liste des operations
symmop=spacegroup.get_symmetry_operations()

########1ere symmetrie=miroir########
print("Symmetrie miroir:")
#on prend les coordonnees du premier atome dans la liste des sites ([1.82778478 -1.05527203  6.00428947])
atomposition=site[1]
print("Les coordonnees initiales sont", atomposition)
#on lui applique la symmetrie miroir qui est la 8eme sur le site Jmol
mirror=symmop[7]
#l'operateur comporte une rotation sans translation
rotation=mirror.rotation_matrix
translation=mirror.translation_vector
print("Matrice de rotation = ",rotation)
print("Vecteur de translation = ",translation)
#on applique la symmetrie miroir a l'atome, les nouvelles coordonnees peuvent aussi etre trouvees avec l'operateur
#de rotation et de translation
new_coord=mirror.operate([1.82778478, -1.05527203, 6.00428947])
print("Les nouvelles coordonnees sont", new_coord)
#on obtient [1.05527203 -1.82778478  6.00428947] pour les coordonnees apres l'operation
plt.figure(figsize = (10,10))
plt.imshow(img.imread('miroir.png'))
plt.show()


########2eme symmetrie=C3-axis########
print("symmetrie C3-axis")
#on prend les coordonnees du premier atome dans la liste des sites ([1.82778478 -1.05527203  6.00428947])
atomposition=site[1]
print("Les coordonnees initiales sont", atomposition)
#on lui applique la symmetrie C3-axis qui est la 5eme sur le site Jmol
C3=symmop[4]
#l'operateur comporte une rotation sans translation
rotation=C3.rotation_matrix
translation=C3.translation_vector
print("Matrice de rotation = ",rotation)
print("Vecteur de translation = ",translation)
#on applique la symmetrie C3-axis a l'atome, les nouvelles coordonnees peuvent aussi etre trouvees avec l'operateur
#de rotation et de translation
new_coord=C3.operate([1.82778478, -1.05527203, 6.00428947])
print("Les nouvelles coordonnees sont", new_coord)
#on obtient [-2.88305681 -1.82778478  6.00428947] pour les coordonnees apres l'operation
plt.figure(figsize = (10,10))
plt.imshow(img.imread('C3axis.png'))
plt.show()


########3eme symmetrie=3baraxis########
print("symmetrie 3baraxis")
#on prend les coordonnees du premier atome dans la liste des sites ([1.82778478 -1.05527203  6.00428947])
atomposition=site[1]
print("Les coordonnees initiales sont", atomposition)
#on lui applique la symmetrie 3baraxis qui est la 18eme sur le site Jmol
baraxis=symmop[17]
#l'operateur comporte une rotation et une translation
rotation=baraxis.rotation_matrix
translation=baraxis.translation_vector
print("Matrice de rotation = ",rotation)
print("Vecteur de translation = ",translation)
#on applique la symmetrie 3baraxis a l'atome, les nouvelles coordonnees peuvent aussi etre trouvees avec l'operateur
#de rotation et de translation
new_coord=baraxis.operate([1.82778478, -1.05527203, 6.00428947])
print("Les nouvelles coordonnees sont", new_coord)
#on obtient [3.54972348  2.16111811 -5.67095614] pour les coordonnees apres l'operation
plt.figure(figsize = (10,10))
plt.imshow(img.imread('3baraxis.png'))
plt.show()
```
